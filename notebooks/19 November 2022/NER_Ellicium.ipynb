```python
from goose3 import Goose

def goose_text_extraction(article_url):
  g = Goose()

  article = g.extract(article_url)
  #article.title
  
  return article.cleaned_text
```

```python
urls = [
    "https://www.nytimes.com/2022/11/09/technology/meta-layoffs-facebook.html",
    "https://www.nytimes.com/2022/11/03/technology/twitter-layoffs-elon-musk.html",
    "https://www.moneycontrol.com/news/business/companies/indian-origin-former-employee-admits-defrauding-apple-of-17-million-9431961.html",
    "https://www.hindustantimes.com/world-news/saudi-aramco-storage-petroleum-facility-hit-by-houthi-attack-causing-fire-101648257221733.html",
    "https://www.thehindu.com/business/Industry/creditsights-finds-errors-in-debt-report-on-indias-adani-group-firms/article65866999.ece",
    "https://www.hindustantimes.com/india-news/data-hacked-from-tata-power-servers-leaked-on-dark-web-101666724211506.html"
]
```

```python
import pandas as pd
  
# Calling DataFrame constructor on list
# with indices and columns specified
df = pd.DataFrame(urls, columns =['URL'])
df
```

```python
df['Text'] = df['URL'].apply(goose_text_extraction)
```

```python
df
```

```python
import spacy
from spacy import displacy

NER = spacy.load('en_core_web_trf')

def NER_Spacy(raw_text, ner_text, ret_org, ret_start, ret_end):

  text_ner = NER(raw_text)
  
  if ner_text:
    return text_ner

  organization = []
  start_char_org = []
  end_char_org = []

  for word in text_ner.ents:
    if word.label_ == 'ORG':
      organization.append(word.text)
      start_char_org.append(word.start_char)
      end_char_org.append(word.end_char)

  if ret_org:
    return organization

  if ret_start:
    return start_char_org

  if ret_end:
    return end_char_org
```

```python
df['NER'] = df['Text'].apply(NER_Spacy, args=(True, False, False, False))
```

```python
df['Organization'] = df['Text'].apply(NER_Spacy, args=(False, True, False, False))
```

```python
df['Start'] = df['Text'].apply(NER_Spacy, args=(False, False, True, False))
```

```python
df['End'] = df['Text'].apply(NER_Spacy, args=(False, False, False, True))
```
